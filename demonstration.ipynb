```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# generate a sample seismic data
t = np.linspace(0, 3.1416, 1000)

# one half of a sine wave
a = np.sin(t)*2

# generate a high frequency noise
noise = np.random.normal(0, 0.1, 1000)

# product the noise with the sine wave
a = a*noise

#plot
plt.plot(t, a)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration vs Time')
plt.show()
```

# windowing

```python
# draw a window of 100 samples and show on the plot
plt.plot(t, a)
plt.plot(t[0:100], a[0:100], 'r')
# draw a box
plt.axvline(x=t[0], color='g')
plt.axvline(x=t[100], color='g')

plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration vs Time')
plt.show()

# move the window by 50 samples
plt.plot(t, a)
plt.plot(t[50:150], a[50:150], 'r')
# draw a box
plt.axvline(x=t[50], color='g')
plt.axvline(x=t[150], color='g')

plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration vs Time')
plt.show()

# move the window by 50 samples
plt.plot(t, a)
plt.plot(t[100:200], a[100:200], 'r')
# draw a box
plt.axvline(x=t[100], color='g')
plt.axvline(x=t[200], color='g')

plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration vs Time')
plt.show()
    
```

```python
# get c3 statistic for the data
def c3(data):
    return np.mean(data**3)/np.mean(data**2)**1.5

c3_stat = c3(a)
print('C3 Statistic: ', c3_stat)
# plot the c3 statistic by a horizontal line
plt.plot(t, a)
plt.axhline(y=c3_stat, color='r', linestyle='-')
plt.xlabel('Time (s)')
plt.ylabel('C3 Statistic')
plt.title('C3 Statistic vs Time')
plt.show()
```

```python
c3_stat = c3(a)
print('C3 Statistic: ', c3_stat)
# plot the c3 statistic by a horizontal line
plt.plot(t, a)
plt.axhline(y=c3_stat, color='r', linestyle='-')
plt.xlabel('Time (s)')
plt.ylabel('C3 Statistic')
plt.title('C3 Statistic vs Time')
plt.show()

# calculate energy of the signal
def energy(data):
    return np.sum(data**2)

energy_stat = energy(a)
print('Energy: ', energy_stat)
# plot the energy statistic by a horizontal line
plt.plot(t, a)
plt.axhline(y=energy_stat, color='r', linestyle='-')
plt.xlabel('Time (s)')
plt.ylabel('Energy')
plt.title('Energy vs Time')
plt.show()
```
